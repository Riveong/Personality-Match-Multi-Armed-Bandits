```python
import random
```

```python
def epsilon_greedy_bandit(dataset, num_arms, epsilon, num_rounds):
    estimated_rewards = [0] * num_arms
    total_rewards = 0
    exploration_count = 0
    exploitation_count = 0

    for _ in range(num_rounds):
        if random.random() < epsilon:
            # Explore: Choose a random arm
            chosen_arm = random.randint(0, num_arms - 1)
            exploration_count += 1
        else:
            # Exploit: Choose the arm with the highest estimated reward
            chosen_arm = max(range(num_arms), key=lambda arm: estimated_rewards[arm])
            exploitation_count += 1

        # Observe the reward (response) for the chosen arm
        reward = dataset.iloc[chosen_arm]['response']
        total_rewards += reward

        # Update the estimated reward for the chosen arm
        estimated_rewards[chosen_arm] += (reward - estimated_rewards[chosen_arm]) / (total_rewards + 1)

    return estimated_rewards, exploration_count, exploitation_count

# Example usage
num_arms = 1000  # Number of items
epsilon = 0.1  # Exploration parameter
num_rounds = 1000  # Number of interactions

estimated_rewards, exploration_count, exploitation_count = epsilon_greedy_bandit(predf, num_arms, epsilon, num_rounds)

# Print results
print(f"Estimated rewards for each arm: {estimated_rewards}")
print(f"Exploration count: {exploration_count}")
print(f"Exploitation count: {exploitation_count}")
```
